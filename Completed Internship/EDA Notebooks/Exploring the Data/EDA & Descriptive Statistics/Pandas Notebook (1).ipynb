Group by (split data into groups based on some criteria and apply function):
df.groupby('A').sum() ; df.groupby(['A', 'B']).sum()

You call .groupby() and pass the name of the column you want to group on. Then, you use ["column"] to specify the column(s) on which you want to perform the actual function.

You can pass a lot more than just a single column name to .groupby() as the first argument. 

```python
data.groupby('Property')['Location'].count()
#data.groupby('age')['type'].count() #ISOLATE EACH TYPE BY DATAFRAMES THAT ARE SEPARATE 
```
